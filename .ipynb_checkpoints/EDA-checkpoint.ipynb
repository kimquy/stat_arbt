```python
import pandas as pd
import numpy as np
import pyarrow.parquet as pq
import glob

files = glob.glob("SOL-USDT/*.parquet")
first = True
for file in files:
    new_df = pq.read_table(file).to_pandas()
    if first:
        sol_df = new_df
        first = False
    else:
        sol_df = pd.concat([sol_df, new_df])
```

```python
files = glob.glob("ADA-USDT/*.parquet")
first = True
for file in files:
    new_df = pq.read_table(file).to_pandas()
    if first:
        ada_df = new_df
        first = False
    else:
        ada_df = pd.concat([ada_df, new_df])
```

```python
files = glob.glob("BNB-USDT/*.parquet")
first = True
for file in files:
    new_df = pq.read_table(file).to_pandas()
    if first:
        bnb_df = new_df
        first = False
    else:
        bnb_df = pd.concat([bnb_df, new_df])
```

```python
files = glob.glob("DOGE-USDT/*.parquet")
first = True
for file in files:
    new_df = pq.read_table(file).to_pandas()
    if first:
        doge_df = new_df
        first = False
    else:
        doge_df = pd.concat([doge_df, new_df])
```

```python
sol_df = sol_df.sort_values(by='timestamp', ascending=True)
bnb_df = bnb_df.sort_values(by='timestamp', ascending=True)
ada_df = ada_df.sort_values(by='timestamp', ascending=True)
doge_df = doge_df.sort_values(by='timestamp', ascending=True)
sol_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
sol_df["close"]=sol_df["close"].astype(float)
sol_df["timestamp"]=sol_df["timestamp"].astype(float)
sol_df["volume"]=sol_df["volume"].astype(float)
sol_df.plot(x="timestamp", y=["close"])

bnb_df["close"]=bnb_df["close"].astype(float)
bnb_df["timestamp"]=bnb_df["timestamp"].astype(float)
bnb_df["volume"]=bnb_df["volume"].astype(float)
bnb_df.plot(x="timestamp", y=["close"])

ada_df["close"]=ada_df["close"].astype(float)
ada_df["timestamp"]=ada_df["timestamp"].astype(float)
ada_df["volume"]=ada_df["volume"].astype(float)
ada_df.plot(x="timestamp", y=["close"])

doge_df["close"]=doge_df["close"].astype(float)
doge_df["timestamp"]=doge_df["timestamp"].astype(float)
doge_df["volume"]=doge_df["volume"].astype(float)
doge_df.plot(x="timestamp", y=["close"])

plt.show()
```

```python
sol_df["usd_volume"]=sol_df["volume"] * sol_df["close"]
bnb_df["usd_volume"]=bnb_df["volume"] * bnb_df["close"]
ada_df["usd_volume"]=ada_df["volume"] * ada_df["close"]
doge_df["usd_volume"]=doge_df["volume"] * doge_df["close"]

x = ['SOL', 'ADA', 'BNB', 'DOGE']
y = [sol_df["usd_volume"].sum(), ada_df["usd_volume"].sum(), bnb_df["usd_volume"].sum(), doge_df["usd_volume"].sum()]
plt.xlabel('Coin')
plt.ylabel('Volume in USD')
plt.title("Total daily volume")
# fig, ax = plt.subplots()
# ax.ticklabel_format(style='plain')

plt.bar(x, y, width = 0.4, align='center')
print(y)
```
